# Cleaning Data

## Bersihkan data minus pada price dan quantity

```python
df_clean = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
```

```python
df_clean.describe()
```

# Hapus Barang Non-Fisik (Ongkir, Manual, dll)

```python
# Menggunakan tanda '~' artinya "YANG TIDAK MENGANDUNG" kata POSTAGE
df_clean = df_clean[~df_clean['StockCode'].astype(str).str.contains('POSTAGE')]
df_clean = df_clean[~df_clean['Description'].astype(str).str.contains('POSTAGE')]
```

```python
df_clean.info()
```

# Membuat Basket dnegan Fokus pada Negara Prancis

```python
basket = (df_clean[df_clean['Country'] == "France"]
        .groupby(['InvoiceNo', 'Description'])['Quantity']
        .sum().unstack().reset_index().fillna(0)
        .set_index('InvoiceNo'))
```

```python
basket.head()
```

# Encoding

```python
def encode_units(x):
    if x <= 0: return 0
    if x >= 1: return 1

basket_sets = basket.applymap(encode_units)
```

# Hasil Basket

```python
# Hapus kolom POSTAGE jika masih ada
if 'POSTAGE' in basket_sets.columns:
    basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
print(f"Jumlah Transaksi (Struk): {basket_sets.shape[0]}")
print(f"Jumlah Jenis Barang: {basket_sets.shape[1]}")
```
